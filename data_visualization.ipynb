```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[8,3,6,1]
```

```python
#Scatter plots
plt.scatter(x,y)
plt.show()
```

```python
#line plots
plt.plot(x,y)
```

```python
#bar plots
categories = ['B','T','S','A','R','M','Y']
values=[3,5,2,8,1,7,4]
plt.bar(categories,values)
```

```python
#pie charts
labels=['Category A','Category B','Category C']
sizes=[30,45,25]
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
```
